```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, PredictionErrorDisplay
```

```python
df = pd.read_csv("fabbisogno calorico.csv", index_col=0)
df.head()
```

```python
x = df[['eta']]
y = df['calorie']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=45)
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(X_test, y_test)
plt.title("Dati primo dello split")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(X_train, y_train)
plt.title("Dati dopo lo split")
plt.grid(True)
plt.show()
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(x, y, color='green', label='Dati reali')
plt.plot(x, model.predict(x), color='salmon', label='Retta di regressione')
plt.xlabel("Et (anni)")
plt.ylabel("Fabbisogno calorico giornaliero (kcal)")
plt.title("Regressione Lineare: Et vs Calorie")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 
print("Mean squad error: %.2f" % mean_squared_error(y_test, y_pred))
print("Coefficiente di detrioramento", r2_score(y_test, y_pred))
print("Coefficiente di detrioramento %.2f" % r2_score(y_test, y_pred))
```

```python
display = PredictionErrorDisplay(y_true=y_test, y_pred=y_pred)
display.plot()
plt.title("Errore di Predizione")
plt.grid(True)
plt.show()
```
