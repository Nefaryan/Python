```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, PredictionErrorDisplay
```

```python
df = pd.read_csv("Student_Performance.csv")

df.head()
```

```python
X = df[['Hours Studied', 'Sleep Hours']]  
y = df['Performance Index']            


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
model = LinearRegression()
model.fit(x_train_scaled, y_train)
```

# Grafici di Studie Hours di train e test

```python
plt.figure(figsize=(12, 7))
# Train 
plt.subplot(2, 2, 1)
plt.scatter(x_train_scaled['Hours Studied'], y_train, color='blue')
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
plt.title('Train Set - Hours Studied vs Performance')
plt.grid(True)

#  Test
plt.subplot(2, 2, 2)
plt.scatter(X_test_scaled['Hours Studied'], y_test, color='orange')
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
plt.title('Test Set - Hours Studied vs Performance')
plt.grid(True)
```

# Grafici di Sleep Hours di train e test

```python
plt.figure(figsize=(12, 7))

# Train 
plt.subplot(2, 2, 1)
plt.scatter(x_train_scaled['Sleep Hours'], y_train, color='green')
plt.xlabel('Sleep Hours')
plt.ylabel('Performance Index')
plt.title('Train Set - Sleep Hours vs Performance')
plt.grid(True)

# Test
plt.subplot(2, 2, 2)
plt.scatter(X_test_scaled['Sleep Hours'], y_test, color='red')
plt.xlabel('Sleep Hours')
plt.ylabel('Performance Index')
plt.title('Test Set - Sleep Hours vs Performance')
plt.grid(True)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
print("Mean squad error: %.2f" % mean_squared_error(y_test, y_pred))
print("Coefficiente di detrioramento", r2_score(y_test, y_pred))
print("Coefficiente di detrioramento %.2f" % r2_score(y_test, y_pred))
```

```python
display = PredictionErrorDisplay(y_true=y_test, y_pred=y_pred)
display.plot()
plt.title("Errore di Predizione")
plt.grid(True)
plt.show()
```
