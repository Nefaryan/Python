```python
# 1. Seelzione delle colonne di title,type e cast
dfCast = df[['title','type','cast']].dropna(subset=['cast'])
```

```python
#Trasforma la colonna cast a riga per ogni singolo attore nei film
dfCast['cast'] = dfCast['cast'].str.split(', ')
actor_df = dfCast.explode('cast')
#Rinominazione della colonna e rimozione di eventuali spazi bainchi
actor_df.rename(columns={'cast': 'actor'}, inplace= True)
actor_df['actor'] = actor_df['actor'].str.strip()
#Raggruppiamo per tipo e attore, poi contiamo i titoli e li mettiamo in una lista
actor_roles = actor_df.groupby(['type', 'actor']).agg(
    count=('title', 'count'),
    titles=('title', list)  
)
# Filtro gli attori con più di una partecipazione e ordino per quelli con più partecipazioni
multi_roles_with_titles = actor_roles[actor_roles['count'] > 1].sort_values(by='count', ascending=False)
```
