```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("calories.csv", index_col=0)
df.head()
```

# Preparazione dei dati per l'analisi

```python
x = df[['kg']]
y = df['calories']

#funzione train_test_split : x,y  test_size indica la percentuale di dati utilizzata per il test, random stat si usa per avere la stessa divisione casuale

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(x_train, y_train)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(x_test, y_test)
plt.show()
```

# Regessione Lineare

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

# Test del Modello

```python
y_pred = model.predict(x_test)
```

# Grafico Plot 

```python
plt.figure(figsize=(12, 7))
plt.scatter(x, y, color='blue', label='Dati reali')
plt.plot(x, model.predict(x), color='red', label='Regressione lineare')
plt.xlabel('Peso (kg)')
plt.ylabel('Calorie bruciate')
plt.title('Regressione Lineare: Peso vs Calorie')
plt.legend()
plt.grid(True)
plt.show()
```

# Valutazione Errore

```python
#MSE (mean squad error) misura dell'errore medio al quadrato
print("Mean squad error: %.2f" % mean_squared_error(y_test, y_pred))
# R2_Score Coefficiente di Determinazione 
print("Coefficiente di detrioramento", r2_score(y_test, y_pred))
print("Coefficiente di detrioramento %.2f" % r2_score(y_test, y_pred))
```

# Grafico errore di predizione

```python
from sklearn.metrics import PredictionErrorDisplay
display = PredictionErrorDisplay(y_true=y_test, y_pred=y_pred)
display.plot()
plt.title("Errore di Predizione")
plt.grid(True)
plt.show()
```
