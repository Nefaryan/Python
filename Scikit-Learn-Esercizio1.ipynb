```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("calories.csv")
df.head()
```

# Preparazione dei dati per l'analisi

```python
x = df[['kg']]
y = df['calories']

#funzione train_test_split : x,y  test_size indica la percentuale di dati utilizzata per il test, random stat si usa per avere la stessa divisione casuale

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# Regessione Lineare

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

# Test del Modello

```python
y_pred = model.predict(x_test)

#Stampa dei risultati
for peso, vero, pred in zip(x_test.values.flatten(), y_test, y_pred):
    print(f"Peso: {peso} kg -> Calorie reali: {vero}, Calorie predette: {pred:.2f}")
```

# Grafico Plot 

```python
plt.figure(figsize=(12, 7))
plt.scatter(x, y, color='blue', label='Dati reali')
plt.plot(x, model.predict(x), color='red', label='Regressione lineare')
plt.xlabel('Peso (kg)')
plt.ylabel('Calorie bruciate')
plt.title('Regressione Lineare: Peso vs Calorie')
plt.legend()
plt.grid(True)
plt.show()
```
